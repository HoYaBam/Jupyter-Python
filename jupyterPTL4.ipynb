```python
import csv
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc

font_path = r'C:/windows/Fonts/gulim.ttc'
font = fm.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

with open('인구현황.csv', 'r', encoding='cp949') as f:
    data = csv.reader(f)
    m = []
    w = []
    
    for row in data:
        if '가수원동' in row[0]:
            for i in range(106, 207):
                m.append(int(row[i]))
                w.append(int(row[i+103]))
    
    plt.title('대전 가수원동 남여 인구 비율')
    plt.scatter(m,w,c=range(101), cmap='jet')
    plt.colorbar()
    plt.plot(range(max(m)),range(max(m)), 'g')
    plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc

font_path = r'C:/windows/Fonts/gulim.ttc'
font = fm.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

f = open('교통카드.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

items = {}
for row in data:
    if row[1] in items:
        items[row[1]][0] += int(row[4])
        items[row[1]][1] += int(row[5])
    else:
        items[row[1]] = [int(row[4]), int(row[5])]

f, axex = plt.subplots(9,3)
f.set_size_inches((10, 40))
plt.subplots_adjust(wspace=0.1, hspace=0.1)

cnt = 0
for key, value in items.items():
    i = cnt // 3
    j = cnt % 3
    axex[i,j].pie(value, labels=['승차','하차'], autopct='%.1f%%')
    axex[i,j].set_title(key)
    cnt += 1

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc

font_path = r'C:/windows/Fonts/gulim.ttc'
font = fm.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

f = open('시간대별.csv', 'r', encoding='cp949')
data = csv.reader(f)

time_ride = []
for row in data:
    time_ride = row
    break


next(data)


items = {}
for row in data:
    if row[1] in items:
        for i in range(4,34,2):
            items[row[1]][time_ride[i]][0] += int(row[i])
            items[row[1]][time_ride[i]][1] += int(row[i+1])
    else:
        time = {}
        for i in range(4,34,2):
            if row[1] in time:
                time[time_ride[i]] += [int(row[i]),int(row[i+1])]
            else:
                time[time_ride[i]] = [int(row[i]),int(row[i+1])]
        items[row[1]] = time
            
            
        
times = []
rides = []
quits = []
for key, value in items.items():
    if key == '2호선':
        for time, num in value.items():
            print('{} : {}'.format(time, num))
            times.append(time)
            rides.append(num[0])
            quits.append(num[1])

print(times)
print(rides)
# print(times)
# print(nums)

plt.plot(times, rides, label = '승차')
plt.plot(times, quits, label = '하차')

plt.legend()
plt.show()
```
