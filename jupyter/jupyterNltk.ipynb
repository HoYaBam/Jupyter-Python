# nltk 사용법

```python
from nltk import word_tokenize
from nltk import text
from nltk.book import *
from nltk import pos_tag
from collections import Counter


with open('speech.txt', 'r') as f:
    data = f.read()

tokens = word_tokenize(data)
text = Text(tokens)


words = pos_tag(text)
# for a in words:
#     if a[1] == 'NN':
#         print(a)

nonus = ['NN', 'NNS', 'NNP', 'NNPS']
nounList = []
for word in words:
    if word[1] in nonus:
#         print(word[0])
        nounList.append(word[0])

counts = Counter(nounList)
# print(counts)

top100 = counts.most_common(100)
print(top100)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

word_cloud = WordCloud(font_path='c://SuncheonR.ttf',background_color="white", max_font_size=60)
cloud = word_cloud.generate_from_frequencies(dict(top100))
plt.figure(figsize=(10,8))
plt.imshow(cloud)
plt.show()
```
