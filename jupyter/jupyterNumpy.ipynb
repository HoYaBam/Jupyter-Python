```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3,'g^')
plt.show()
```

```python
x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)
mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1+mask2]
y = y[mask1+mask2]

size = np.random.rand(100) * 1000
plt.scatter(x,y, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
