# CSV 파일 요약 실습
## 9세 이상 24세 이하인 청소년 인구가 많은 순으로 10개 지역을 찾아보고
## 막대그래프로 작성하시오
## 28살부터 35살 까지

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
f = open('인구현황.csv', 'r', encoding='cp949')
from matplotlib import font_manager, rc
font_path = r'C:/windows/Fonts/gulim.ttc'
font = font_manager.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)
data = csv.reader(f)
next(data)
header = next(data)

data = list(data)

away = []
city_name_value = {}

for row in data:
    if '' in row[0]:
        if row[0].count('0',0, len(row[0])) < 6:
            away.append(np.sum(np.array(row[12:28],dtype=int)))

away.sort(reverse=True)
        
for i in range(10):
    for row in data:
        if np.sum(np.array(row[12:28],dtype=int)) == away[i]:
            city_name_value[row[0]] = away[i]
            
count = 1
ryxhd = []
ryxhdN = []
for key , value in city_name_value.items():
    print("청소년이 가장 많은 지역 {}위는 {} 이며 인구수는 {} 입니다.".format(count,key[0: key.find('(')],value))
    count += 1
    ryxhd.append(key[0: key.find('(')])
    ryxhdN.append(value)
    
plt.figure(figsize=(40,20))
plt.bar(ryxhd, ryxhdN)
plt.show()
```
