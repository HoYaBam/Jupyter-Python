```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import re
import csv

driver = webdriver.Chrome(executable_path='./chromedriver.exe')
url = "https://www.goodhosrank.com/hospital/hospital_list.php?stationid=&linenum=&orderby=&hashtag=&orderby=&hostime2=&searchkeyword=&addcode=30&hostime=&gubun=%EC%A2%85%ED%95%A9%EB%B3%91%EC%9B%90"
driver.get(url)
time.sleep(2)

csvfile = open('병원진료과목1.csv', 'wt+', newline="")
csvfilew = csv.writer(csvfile)

count = driver.find_elements_by_xpath('//*[@id="body_main"]/table/tbody/tr')

for i in range(2,len(count),2):
    subjectList = []
    divCount = driver.find_elements_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div'.format(i))
    if len(divCount) == 4:
        title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
        subjectList.append(title)
        subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[4]'.format(i)).text)
        csvfilew.writerow(subjectList)
    elif len(divCount) == 5:
        title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
        subjectList.append(title)
        subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[5]'.format(i)).text)
        csvfilew.writerow(subjectList)
    else:
        pass

search_box = driver.find_element_by_xpath('//*[@id="body_main"]/div[6]/div[4]/form/select')
search_box.send_keys('일반병원')
time.sleep(1)

for p in range(4):
    for i in range(2,len(count),2):
        subjectList = []
        divCount = driver.find_elements_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div'.format(i))
        if len(divCount) == 4:
            title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
            subjectList.append(title)
            subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[4]'.format(i)).text)
            csvfilew.writerow(subjectList)
        elif len(divCount) == 5:
            title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
            subjectList.append(title)
            subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[5]'.format(i)).text)
            csvfilew.writerow(subjectList)
        else:
            pass
    
    nextPage = driver.find_element_by_xpath('//*[@id="body_main"]/div[8]/table/tbody/tr/td[3]/a')
    nextPage.click()
    time.sleep(2)
    
search_box = driver.find_element_by_xpath('//*[@id="body_main"]/div[6]/div[4]/form/select')
search_box.send_keys('치과병원')
time.sleep(1)

for i in range(2,len(count),2):
    subjectList = []
    divCount = driver.find_elements_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div'.format(i))
    if len(divCount) == 4:
        title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
        subjectList.append(title)
        subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[4]'.format(i)).text)
        csvfilew.writerow(subjectList)
    elif len(divCount) == 5:
        title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
        subjectList.append(title)
        subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[5]'.format(i)).text)
        csvfilew.writerow(subjectList)
    else:
        pass

search_box = driver.find_element_by_xpath('//*[@id="body_main"]/div[6]/div[4]/form/select')
search_box.send_keys('한방병원')
time.sleep(1)

for p in range(1):
    for i in range(2,len(count),2):
        subjectList = []
        divCount = driver.find_elements_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div'.format(i))
        if len(divCount) == 4:
            title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
            subjectList.append(title)
            subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[4]'.format(i)).text)
            csvfilew.writerow(subjectList)
        elif len(divCount) == 5:
            title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
            subjectList.append(title)
            subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[5]'.format(i)).text)
            csvfilew.writerow(subjectList)
        else:
            pass
    nextPage = driver.find_element_by_xpath('//*[@id="body_main"]/div[8]/table/tbody/tr/td[3]/a')
    nextPage.click()
    time.sleep(2)
    
search_box = driver.find_element_by_xpath('//*[@id="body_main"]/div[6]/div[4]/form/select')
search_box.send_keys('요양병원')
time.sleep(1)

for p in range(5):
    for i in range(2,len(count),2):
        subjectList = []
        divCount = driver.find_elements_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div'.format(i))
        if len(divCount) == 4:
            title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
            subjectList.append(title)
            subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[4]'.format(i)).text)
            csvfilew.writerow(subjectList)
        elif len(divCount) == 5:
            title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
            subjectList.append(title)
            subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[5]'.format(i)).text)
            csvfilew.writerow(subjectList)
        else:
            pass
    nextPage = driver.find_element_by_xpath('//*[@id="body_main"]/div[8]/table/tbody/tr/td[3]/a')
    nextPage.click()
    time.sleep(2)
    
search_box = driver.find_element_by_xpath('//*[@id="body_main"]/div[6]/div[4]/form/select')
search_box.send_keys('기타병원')
time.sleep(1)
    
for i in range(2,len(count),2):
    subjectList = []
    divCount = driver.find_elements_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div'.format(i))
    if len(divCount) == 4:
        title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
        subjectList.append(title)
        subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[4]'.format(i)).text)
        csvfilew.writerow(subjectList)
    elif len(divCount) == 5:
        title = driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[1]/a'.format(i)).text
        subjectList.append(title)
        subjectList.append(driver.find_element_by_xpath('//*[@id="body_main"]/table/tbody/tr[{}]/td[2]/div[5]'.format(i)).text)
        csvfilew.writerow(subjectList)
    else:
        pass


print("끝")
csvfile.close()
```
